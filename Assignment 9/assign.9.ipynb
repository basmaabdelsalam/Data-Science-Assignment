```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
df = pd.read_csv('50ulke.csv')
```

```python
df.head()
```

# Distribution Plots 
## distplot()

```python
sns.distplot(df['Growth'], kde=False, color='m')
```

```python
sns.distplot(df['National Income'], kde=False, color='m')
```

```python
sns.distplot(df['Per person'], kde=False, color='m')
```

## kdeplot()

```python
sns.kdeplot(df['National Income'], shade=True, color='m')
```

```python
sns.kdeplot(df['Growth'], shade=True, color='m')
```

```python
sns.kdeplot(df['Per person'], shade=True, color='m')
```

## Jointplot()

```python
sns.jointplot(x='Growth', y='Per person', data=df, color='m', height=7, ratio=5)
```

```python
sns.jointplot(x='Growth', y='National Income', data=df, color='m', height=7, ratio=5)
```

```python
sns.jointplot(x='National Income', y='Per person', data=df, color='m', height=7, ratio=5)
```

```python
sns.jointplot(x='Growth', y='National Income', data=df, color='m',kind='kde', height=7, ratio=5)
```

```python
sns.jointplot(x='Growth', y='Per person', data=df,kind='kde', color='m', height=7, ratio=5)
```

```python
sns.jointplot(x='National Income', y='Per person', data=df, color='m',kind='kde', height=7, ratio=5)
```
